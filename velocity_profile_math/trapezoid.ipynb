```python
import scipy.integrate as it
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Distance to cover
D = 5
```

```python
# Hyperparameters
a = 0.8
v_max = min(D/2, 1)
```

```python
# Time computation

tp1 = v_max/a
tp2 = (a*D - (v_max**2))/(a*v_max)
tp3 = v_max/a

T = tp1 + tp2 + tp3
```

```python
# Time steps
t_steps = np.arange(0, T, 0.04)
```

```python
# velocity function

def v(t):
    if t <= tp1:
        return a*t
    elif t <= tp1 + tp2:
        return v_max
    else:
        return v_max - a*(t - tp1 - tp2)
```

```python
vel = []

for t in t_steps:
    vel.append(v(t))
```

```python
plt.plot(t_steps, vel)
plt.plot((tp1, tp1), (0, v_max), '--')
plt.plot((tp1+tp2, tp1+tp2), (0, v_max), '--')
plt.plot((tp1+tp2+tp3, tp1+tp2+tp3), (0, v_max), '--')
plt.title("Velocity-Time Graph")
plt.show()
```

```python
p = it.cumulative_trapezoid(vel, t_steps)
```

```python
plt.plot(t_steps[1:], p)
plt.plot((0, T), (D, D), '--')
plt.title("Position-Time Graph")
plt.show()
```

```python
print(f"Final Position: {p[-1]}")
```
