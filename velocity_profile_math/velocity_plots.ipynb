```python
import json
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as it
```

```python
with open('constant_telemetry.json', 'r') as f:
    const_vels = json.load(f)['state']
```

```python
x_vel_const = [v['velX'] for v in const_vels]
y_vel_const = [v['velY'] for v in const_vels]
z_vel_const = [v['velZ'] for v in const_vels]

plt.plot(range(len(x_vel_const)), x_vel_const)
# plt.plot(range(len(y_vel_const)), y_vel_const)

plt.legend(['X-Velocity', 'Y-Velocity'])
plt.show()
```

```python
x_pos_const = it.cumulative_trapezoid(x_vel_const, [i*0.1 for i in range(len(x_vel_const))])
y_pos_const = it.cumulative_trapezoid(y_vel_const, [i*0.1 for i in range(len(y_vel_const))])

plt.plot(range(len(x_pos_const)), x_pos_const)
# plt.plot(range(len(y_pos_const)), y_pos_const)

plt.legend(['X-Position', 'Y-Position'])
plt.show()
```
